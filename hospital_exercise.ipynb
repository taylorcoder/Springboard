# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## Analysis of initial observations based on the plot.

+ While there does appear to be a weak negative correlation between rate of readmissions and number of discharges, it does not appear to be substantial. In particular, the vast majority of the data seems to observe very little change in the excess readmission rate as number of discharges 
+ It appears to be true that for lower number of discharges, there is a greater incidence of excess rate of readmissions (area shaded red)

```python
# Analyze statistics for hospitals with number of discharges < 100

df_small_discharges = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] < 100]

mean_excess_small = df_small_discharges['Excess Readmission Ratio'].mean()

print('The mean excess readmission rate for hospitals with number of discharges < 100 is:', mean_excess_small)

excess_greater_small = 100*(df_small_discharges['Excess Readmission Ratio'] > 1).sum() / (df_small_discharges['Excess Readmission Ratio'] > 1).count()

print('The percent of these hospitals with excess readmission ratio > 1 is:', excess_greater_small)

# Analyze statistics for hospitals with number of discharges > 1000

df_large_discharges = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] > 1000]

mean_excess_large = df_large_discharges['Excess Readmission Ratio'].mean()
print('The mean excess readmission rate for hospitals with number of discharges > 1000 is:', mean_excess_large)

excess_greater_large = 100* (df_large_discharges['Excess Readmission Ratio'] > 1).sum() / (df_large_discharges['Excess Readmission Ratio'] > 1).count()
print('The percent of these hospitals with excess readmission ratio > 1 is:', excess_greater_large)
```

## Analysis of initial statistics

+ I agree with all of the above statistics, except for the percent of hospitals with number of discharges < 100 that experienced an excess readmission ratio > 1. For this, I obtained a value of 59%, not 63%.

```python
### Hypothesis test on correlation

# Define a function to return the pearson correlation coefficient between two columns

def pearson_r(x,y):

    corr_mat = np.corrcoef(x,y)

    return corr_mat[0,1]

# Store the columns of interest in numpy arrays

discharges = np.array(clean_hospital_read_df['Number of Discharges'][81:-3])
excess_readmission_ratios = np.array(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

# Compute the observed correlation coefficient

r_observed = pearson_r(discharges, excess_readmission_ratios)

# Initialize an empty array to store permutation replicates

perm_replicates = np.empty(10000)

# Draw replicates for the correlation coefficient on permuted discharges, holding readmission ratios fixed

for i in range(10000):
    
    discharges_permuted = np.random.permutation(discharges)

    perm_replicates[i] = pearson_r(discharges_permuted, excess_readmission_ratios)
    
# Compute the p-value

p = np.sum(perm_replicates<=r_observed)/len(perm_replicates)
print('p-val =', p)
print('The observed correlation coefficient for this dataset is:', r_observed)
```

## Analysis of hypothesis test

+ We start with the null hypothesis that number of discharges and excess readmission ratio are completely uncorrelated. Obtaining a very small p-value (reported here as p = 0), we can conclude that the null hypothesis is false. It appears that number of discharges appears to be correlated with excess readmission ratio. However, although the test reported statistical significance, it does not necessarily imply practical significance! In this case, the observed value for the pearson correlation coefficient was -0.09, a very weak negative correlation. 
+ For pratical purposes, making the above recommendations make sense, since the relationship between number of discharges and readmission ratio is most notable for very small and very large hospitals. However, we might be more interested in exploring other variables that are more strongly correlated with the excess readmission rate, such as the socioeconomic status of the region surrounding the hospitals.

## Analysis of scatterplot

+ While this scatterplot does show the trend between discharges and excess readmission ratio for very small and very large hospitals, the vast majority of the data is overlapping in a cluster between 0 and 1000 discharges, and between a readmission ratio of 0.8 to 1.2. This makes it difficult for us to determine the trend between the two variables in this region. To clearly display the trend, we will display in a scatterplot only hospitals with either a large (>1.2) or small (<0.8) excess readmission ratio, and display the corresponding linear regression line.

```python
df_extreme_discharges = clean_hospital_read_df[(clean_hospital_read_df['Excess Readmission Ratio'] <= 0.8) | 
                                               (clean_hospital_read_df['Excess Readmission Ratio'] >=1.2)]

x = [a for a in df_extreme_discharges['Number of Discharges'][:-3]]
y = list(df_extreme_discharges['Excess Readmission Ratio'][:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions (extreme cases)', fontsize=14)

slope, intercept = np.polyfit(x,y,1)
x_line = np.array([0,3000])
y_line = slope * x_line + intercept
plt.plot(x_line, y_line)

ax.grid(True)
fig.tight_layout()
```

## Analysis of new plot

+ By focusing on hospitals with more extreme readmissions, we can see a much more apparent trend between number of discharges and excess readmission ratio
